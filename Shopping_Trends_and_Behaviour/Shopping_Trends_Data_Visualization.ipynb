```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Q1) What is the distribution of customer ages in your dataset?

print(f'Minimum Age: {trends_data.Age.min()}, Maimum Age: {trends_data.Age.max()}')

sns.set_style("whitegrid")

plt.title("Distribution of Customer Age")
plt.hist(trends_data.Age);
plt.show()
```

```python
trends_data['Item Purchased'].value_counts().index
```

```python
# Q2) What is the total number of each item purchased?

plt.figure(figsize=(12,6))

plt.title('Total Number of Each Purchased Item')
plt.xlabel('Item Purchased')
plt.ylabel('Total Number')
plt.bar(trends_data['Item Purchased'].value_counts().index, trends_data['Item Purchased'].value_counts());
plt.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Q3) How does the gender distribution look among your customers?

gender_counts = trends_data['Gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', explode=(0.05,0))
plt.legend()
plt.show()
```

```python
# Q4) What are the total purchase amounts for each category of items?

category_purchase = trends_data[['Category', 'Purchase Amount (USD)']].copy().groupby('Category').sum().reset_index()
category_purchase
```

```python
plt.title('Total Purchase Amounts for Each Category')
plt.bar(category_purchase.Category, category_purchase['Purchase Amount (USD)'])
plt.xlabel('Category')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
# Q5) How do purchases vary across different seasons?

season_purchase = trends_data[['Season', 'Purchase Amount (USD)']].copy().groupby('Season').sum().reset_index()
season_purchase
```

```python
plt.title('Total Purchase Amounts for Each Season')
plt.bar(season_purchase.Season, category_purchase['Purchase Amount (USD)'])
plt.xlabel('Season')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
# Q6) Can you visualize the total purchase amounts for different locations?

location_purchase = trends_data[['Location', 'Purchase Amount (USD)']].copy().groupby('Location').sum().reset_index()
```

```python
location_purchase = location_purchase.sort_values(by='Purchase Amount (USD)')

# matplotlib.rcParams.update(matplotlib.rcParamsDefault)
plt.figure(figsize=(7,10))
plt.title('Total Purchase Amounts for Each Location')
# plt.barh(location_purchase.Location, location_purchase['Purchase Amount (USD)'])
# plt.spines[['right', 'top', 'bottom']].set_visible(False)
plt.gca().spines[['right', 'top', 'bottom']].set_visible(False)
plt.gca().get_xaxis().set_visible(False)
plt.bar_label(plt.barh(location_purchase.Location, location_purchase['Purchase Amount (USD)'], height=0.6))
plt.xlabel('Location')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
# Q7) Is there any correlation between the review ratings and the purchase amounts?

colors = sns.color_palette('viridis', n_colors=len(trends_data['Review Rating'].unique()))

plt.figure(figsize=(8, 5))
plt.title('Correlation between Review Ratings and Purchase Amounts')

for rating, rating_color in zip(sorted(trends_data['Review Rating'].unique()), colors):
  subset = trends_data[trends_data['Review Rating'] == rating]
  plt.scatter(subset['Purchase Amount (USD)'], subset['Review Rating'], alpha=0.4, color=rating_color)

plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Review Rating')
plt.show()

# Graph shows no correlation between the review ratings and the purchase amounts
```

```python
# Q8) What is the distribution of customers based on their subscription status?

subscription = trends_data['Subscription Status'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(subscription, labels=subscription.index, autopct='%1.1f%%', explode=(0.05,0))
plt.legend()
plt.show()
```

```python
# Q9) How do customers prefer to pay? Visualize the distribution of payment methods.

plt.figure(figsize=(7.5,5))

plt.title('Total Number of Each Purchased Item')
plt.xlabel('Payment Method')
plt.ylabel('Number of Payments by Payment Method')
plt.bar(trends_data['Payment Method'].value_counts().index, trends_data['Payment Method'].value_counts());
plt.tight_layout()
plt.show()
```

```python
trends_data['Payment Method'].value_counts()
```

```python
# Q10) Is there a noticeable impact of discounts on the purchase amounts?

discount_purchases = trends_data[['Discount Applied', 'Purchase Amount (USD)']].groupby('Discount Applied').mean().reset_index()

plt.bar(discount_purchases['Discount Applied'], discount_purchases['Purchase Amount (USD)'])
plt.show()
```

```python
# Q11) What is the distribution of the frequency of purchases among customers?

plt.title('Distribution of Frequency of Purchases')
plt.barh(trends_data['Frequency of Purchases'].value_counts().index, trends_data['Frequency of Purchases'].value_counts())
plt.xlabel('Number of Customers')
plt.ylabel('Frequency Categories')
plt.show()
```

```python
# Q12) Category distribution for each location

top_locations = trends_data['Location'].value_counts().head(5).index

# Create a subplot grid for each location
fig, axes = plt.subplots(5, 1, figsize=(10, 15))

# Iterate through the top locations and create category distribution plots with different colors
for i, location in enumerate(top_locations):
    location_data = trends_data[trends_data['Location'] == location]

    # Count product categories in this location
    category_counts = location_data['Category'].value_counts()

    # Create a bar plot for the category distribution with different colors
    ax = axes[i]
    category_counts.plot(kind='bar', ax=ax, color=['#98FB98', '#FFE5CC', '#7f7f7f', '#CCE5FF'])
    ax.set_title(f"Categories in {location}")
    ax.set_xlabel("Category")
    ax.set_ylabel("Count")
    ax.set_xticklabels(category_counts.index, rotation=45)
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust subplot layout for a clean appearance
plt.tight_layout()

# Display the visualizations
plt.show()
```

```python
# Q13) What is the category distribution by age group?

age_groups = pd.cut(x=trends_data['Age'], bins=[15,25,35,45,55,65,75])

sns.countplot(x=trends_data['Category'], hue=age_groups, palette="Spectral")
plt.show()
```

```python
# Q14) What is the average purchase amount by age?

plt.figure(figsize=(10,6))
plt.title("Average purchase amount by age", fontsize = 20)
# plt.ylim([40,100])
sns.lineplot(x=trends_data.Age, y=trends_data['Purchase Amount (USD)'], marker="o", hue=trends_data.Gender)
plt.xlabel("Age", fontsize = 15)
plt.ylabel("Purchase Amount (USD)", fontsize = 15)
```
